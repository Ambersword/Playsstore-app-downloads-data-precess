```python
from pandas.io.pytables import IndexCol
import pandas as pd
import numpy as np
from numpy import nan
from google.colab import files
uploaded = files.upload()
```

```python
from pandas.tseries.frequencies import unique
import pandas as pd
import io
playstore_app= pd.read_csv(io.BytesIO(uploaded['playstore_apps.csv']))
playstore_app=playstore_app.set_index('App',drop=True,inplace=False)
playstore_app
```

```python
playstore_app.info
```

```python
playstore_app.describe()
```

```python
playstore_app.drop_duplicates(keep=False,inplace=True)
playstore_app.describe()
```

```python
playstore_app['Rating']=playstore_app['Rating'].fillna(0)
playstore_app['Rating'].unique()
```

```python
print(playstore_app.index.unique())
print(playstore_app["Reviews"].unique())
print(playstore_app["Size"].unique())
print(playstore_app["Type"].unique())
print(playstore_app["Price"].unique())
print(playstore_app["Content Rating"].unique())
print(playstore_app["Genres"].unique())
```

```python
playstore_app=playstore_app.dropna(axis=0,subset=['Type'])
playstore_app['Type'].unique()
```

```python
playstore_app=playstore_app.dropna(axis=0,subset=['Content Rating'])
playstore_app['Content Rating'].unique()
```

```python
playstore_app['Current Ver']=playstore_app['Current Ver'].replace("","nan")
```

```python
playstore_app=playstore_app.dropna(axis=0,subset=['Android Ver'])
playstore_app['Android Ver'].unique()
```

```python
playstore_app.info()
```

```python
from IPython.utils.sysinfo import encoding
playstore_app.to_csv('cleaned_app.csv',encoding='utf-8')
files.download('cleaned_app.csv')
```
